```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_tracks=pd.read_csv("tracks.csv")
df_tracks.head()
```

```python
#top 10 most popular songs greater than 90 in spotify
most_pop=df_tracks.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_pop[:10]
#inplace=False:-coz i don't want to change the structure of the table
```

```python
df_tracks.reset_index(inplace=True)  # bring 'release_date' back as column
df_tracks['release_date'] = pd.to_datetime(df_tracks['release_date'], errors='coerce')#errors='coerce': if any value is not a proper date (like just "1922"), 
#it becomes NaT (missing date).
df_tracks.set_index('release_date', inplace=True)
df_tracks.head()
```

```python
#check the artist at the 18th row in the dataset
df_tracks[['artists']].iloc[18]
```

```python
#to convert milli seconds to seconds in duration col
df_tracks['duration'] = df_tracks['duration_ms'] // 1000
```

```python
# Drop all non-numeric columns before correlation
corr = df_tracks.drop(['id', 'name', 'artists', 'id_artists', 'key', 'mode', 'explicit', 'release_date'], axis=1).corr(method="pearson")

# Plot heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(corr, annot=True, cmap='inferno', center=0, linewidths=1, linecolor='black', fmt=".1g")
plt.title("Correlation Heatmap")
plt.xticks(rotation=90)
plt.show()
#sns.heatmap(...): creates a heatmap (colored grid).
#corr: the correlation matrix to display.
#annot=True: show the actual numbers on each square.
#cmap='inferno': sets the color theme (Inferno = black-red-orange-yellow).
#center=0: center of the color scale is 0 (neutral correlation).
#linewidths=1: lines between boxes are 1 unit thick.
#linecolor='black': the border lines are black.
#fmt=".1g": format numbers with 1 significant digit.
```

```python
#to Convert the text in 'release_date' column into proper datetime format.
df_tracks['dates'] = pd.to_datetime(df_tracks['release_date'], format='mixed', errors='coerce')
df_tracks['year'] = df_tracks['dates'].dt.year
print(df_tracks['year'])
#format='mixed': allows different date formats in the column.
#errors='coerce': if a value still can’t be converted, it becomes NaT (safe).
#.dt.year: extracts year from valid dates.
```

```python
#totl num of songs each yr since 1920
sns.displot(df_tracks['year'].dropna(), discrete=True, aspect=2, height=5, kind="hist")
plt.title("Number of Songs per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
#df_tracks['year'].dropna() → removes any missing years (from bad date conversions).
#discrete=True → makes x-axis show separate year bars.
#aspect=2, height=5 → controls width & height of the plot.
```

```python
#duration of the songs over the year
# Group by year and calculate average duration
avg_duration = df_tracks.groupby('year')['duration'].mean().reset_index()

# Plot
plt.figure(figsize=(18, 7))
sns.barplot(x='year', y='duration', data=avg_duration, errorbar=None)
plt.title("Year vs Average Song Duration")
plt.xlabel("Year")
plt.ylabel("Duration (seconds)")
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate average duration per year
df_tracks['year'] = df_tracks['dates'].dt.year
avg_duration = df_tracks.groupby('year')['duration'].mean().reset_index()

# Plot
sns.set_style("whitegrid")
plt.figure(figsize=(10, 5))
sns.lineplot(data=avg_duration, x='year', y='duration').set(title="Year vs Average Duration")
plt.xlabel("Year")
plt.ylabel("Duration (seconds)")
plt.xticks(rotation=90)
plt.show()
#.dt.year → extracts just the year (e.g., from 2001-05-04 it gets 2001).
#df_tracks['dates'] → accesses the existing column with full dates.
#df_tracks.groupby('year') → groups all rows with the same year.
#['duration'] → picks only the duration column from each group.
#.mean() → calculates the average song duration per year.
#.reset_index() → turns the result into a neat DataFrame (not Series).
```

```python
df_genre=pd.read_csv("SpotifyFeatures.csv")
df_genre.head()
```
