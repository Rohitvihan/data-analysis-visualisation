```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('item.csv')

# Replace null values in quantity column with mean
df['quantity'].fillna(df['quantity'].mean(), inplace=True)
df

# Replace null values in price column with median
df['price'].fillna(df['price'].median(), inplace=True)

# Replace null values in bought column with standard deviation
df['bought'].fillna(df['bought'].std(), inplace=True)

# Replace null values in forenoon column with minimum
df['forenoon'].fillna(df['forenoon'].min(), inplace=True)

# Replace null values in afternoon column with maximum
df['afternoon'].fillna(df['afternoon'].max(), inplace=True)

df
```
