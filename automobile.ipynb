```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


automobile_data = pd.read_csv('Automobile_data.csv')

# i) Distribution of two and four-door cars with respect to the type of fuel they use
sns.countplot(x='num-of-doors', hue='fuel-type', data=automobile_data)
plt.title('Distribution of Two and Four-door Cars by Fuel Type')
plt.xlabel('Number of Doors')
plt.ylabel('Count')
plt.show()


```

```python
# ii) Distribution of cars of different body styles with respect to the type of fuel they use
sns.countplot(x='body-style', hue='fuel-type', data=automobile_data)
plt.title('Distribution of Cars by Body Style and Fuel Type')
plt.xlabel('Body Style')
plt.ylabel('Count')
plt.show()

```

```python
# iii) Total number of each type of body style cars categorized by fuel type
pd.crosstab(automobile_data['body-style'], automobile_data['fuel-type']).plot(kind='bar', stacked=True)
plt.title('Total Number of Each Body Style by Fuel Type')
plt.xlabel('Body Style')
plt.ylabel('Count')
plt.show()

```

```python
# iv) Horsepower of each fuel type with reference to the type of drive wheel present in cars
# Ensure 'horsepower' is numeric and handle missing values
automobile_data['horsepower'] = pd.to_numeric(automobile_data['horsepower'], errors='coerce')
automobile_data = automobile_data.dropna(subset=['horsepower'])

# Create the box plot
sns.boxplot(x='fuel-type', y='horsepower', hue='drive-wheels', data=automobile_data)
plt.title('Horsepower of Each Fuel Type by Drive Wheel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Horsepower')
plt.show()
```
